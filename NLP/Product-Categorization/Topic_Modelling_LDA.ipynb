```python
## Text Preprocessing
import re
def tokens(words):
    words = re.sub("[^a-zA-Z]"," ",words)
    text = words.lower().split()
    return " ".join(text)
```

```python
df_clean['title'] = df_clean['title'].apply(tokens)
df_clean.head()
```

```python
## Stop words
import nltk 
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def pre_preprocess(text):

    # Tokenize the text into words
    word_tokens = word_tokenize(text)

    # Create a set of stopwords
    stop_words = set(stopwords.words('english'))

    # Filter out the stopwords
    filtered_sentence = [word for word in word_tokens if word.lower() not in stop_words]

    return " ".join(filtered_sentence)
```

```python
# test stop words
text  ="get in the game i ultra suit in combo with different straps watch cover wireless charger bt calling heart fitness tracker multicolor ultra watch"

print(pre_preprocess(text))
```

```python
df_clean['title'] = df_clean['title'].apply(pre_preprocess)
df_clean.head()
```

# LDA with Scikit Learn

```python
!pip3 install scikit-learn
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vectorizer = CountVectorizer(analyzer='word',       
                             min_df=3,                       
                             stop_words='english',             
                             lowercase=True,                   
                             token_pattern='[a-zA-Z0-9]{3,}',  
                             max_features=5000,          
                            )

vectorized = vectorizer.fit_transform(df_clean['title'])
```

```python
model_lda = LatentDirichletAllocation(n_components=6, # Number of topics
                                      learning_method='online',
                                      random_state=0,       
                                      n_jobs = -1 
                                     )

lda_output = model_lda.fit_transform(vectorized)
```

## The above topic cluster is well separated and thus it is a good looking topic model 

```python
import numpy as np
vocab = vectorizer.get_feature_names_out()

n_top_words = 5
k = 6

topic_words = {}
for topic, comp in enumerate(model_lda.components_):

    word_idx = np.argsort(comp)[::-1][:n_top_words] 

    # store the words most relevant to the topic
    topic_words[topic] = [vocab[i] for i in word_idx]
```

```python
for topic, words in topic_words.items():
    print('Topic: %d' % topic)
    print('  %s' % ', '.join(words))
```

```python
topic_distribution = model_lda.transform(vectorized)
print(topic_distribution)
```

```python
classified_texts = []
for text in topic_distribution:
    topic_index = text.argmax()
    classified_texts.append( topic_index)
print(classified_texts)
df_topic_cluster = pd.DataFrame({'topic_indices': classified_texts})
# classified_texts = pd.DataFrame({'topic_indices': topic_indices})
df_topic_cluster.count()
```

```python
# Assuming df is your DataFrame and topic_indices is your list of topic indices
final_df = pd.merge(df_topic_cluster, df_clean, left_index=True, right_index=True)
final_df.head()
```

```python
### Checking for topic distribtion for any imbalance in data
final_df['topic_indices'].value_counts()
```
